```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
sns.set_style("darkgrid")
```

```python
# Read And EXplore the data
train = pd.read_csv("../Dataset/titanic.csv")
train.head(n=10)
```

```python
sns.set(rc={'figure.figsize':(10, 8)})
# Correlation matrix between numerical values (SibSp Parch Age and Fare values) and Survived 
#Only Fare feature seems to have a significative correlation with the survival probability.
sns.heatmap(train[['Survived', 'SibSp', 'Parch', 'Age', 'Pclass', 'Fare']].corr(),
            annot=True,
            fmt='.2f',
            cmap='coolwarm')
plt.show()
```

```python
#age vs Survived
#sort the ages into logical categories
train["Age"] = train["Age"].fillna(-0.5)
# test["Age"] = test["Age"].fillna(-0.5)
bins = [-1, 0, 5, 12, 18, 24, 35, 60, np.inf]
labels = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young-Adult', 'Adult', 'Senior']
train['AgeGroup'] = pd.cut(train["Age"], bins, labels = labels)
# test['AgeGroup'] = pd.cut(test["Age"], bins, labels = labels)

#draw a bar plot of Age vs. survival
sns.barplot(x="AgeGroup", y="Survived", data=train)
plt.show()
```

```python
#Age Vs Survived
age_survived_plt = sns.FacetGrid(train, col='Survived', height=4, aspect=1)
age_survived_plt.map(sns.distplot, 'Age')

plt.show()
#So, It's look like age distributions are not the same in the survived and
#not survived subpopulations. Indeed, there is a peak corresponding to young passengers, 
#that have survived. We also see that passengers between 60-80 have less survived. 
#So, even if "Age" is not correlated with "Survived", we can see that there is age 
#categories of passengers that of have more or less chance to survive.
```

```python
#Age vs Sex VS Survived

age_sex_survived_plt = sns.FacetGrid(train , hue='Survived' , aspect=4 , row = 'Sex' )
age_sex_survived_plt.map( sns.kdeplot, 'Age' , shade= True )
age_sex_survived_plt.set( xlim=( 0 , train['Age'].max() ) )
age_sex_survived_plt.add_legend()
plt.show()
#Again we see that aged passengers between 65-80 have less survived.
```

```python
survived = 'survived'
not_survived = 'not survived'
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = train[train['Sex']=='female']
men = train[train['Sex']=='male']
ax = sns.distplot(women[women['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[0], kde =False)
ax = sns.distplot(women[women['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[0], kde =False)
ax.legend()
ax.set_title('Female')
ax = sns.distplot(men[men['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[1], kde = False)
ax = sns.distplot(men[men['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[1], kde = False)
ax.legend()
_ = ax.set_title('Male')

#You can see that men have a high probability of survival when they are between 18 and 30 years old
#For women the survival chances are higher between 14 and 40.
```

```python
#draw a bar plot of survival by sex
#It is clearly obvious that Male have less chance to survive than Female. 
#This is heavily an important feature for our prediction task.
sns.barplot(x="Sex", y="Survived", data=train)

#print percentages of females vs. males that survive
print("Percentage of females who survived:", train["Survived"][train["Sex"] == 'female'].value_counts(normalize = True)[1]*100)

print("Percentage of males who survived:", train["Survived"][train["Sex"] == 'male'].value_counts(normalize = True)[1]*100)
```

```python
#draw a bar plot for SibSp vs. survival
sns.barplot(x="SibSp", y="Survived", data=train)

#It seems that passengers having a lot of siblings/spouses have less chance to survive. 
#Single passengers (0 SibSP) or with two other persons (SibSP 1 or 2) have more chance to survive
print("Percentage of SibSp = 0 who survived:", train["Survived"][train["SibSp"] == 0].value_counts(normalize = True)[1]*100)

print("Percentage of SibSp = 1 who survived:", train["Survived"][train["SibSp"] == 1].value_counts(normalize = True)[1]*100)

print("Percentage of SibSp = 2 who survived:", train["Survived"][train["SibSp"] == 2].value_counts(normalize = True)[1]*100)
```

```python
#SibSp VS Survived
sibsp_survived_plt = sns.FacetGrid(train, col='Survived', height=4, aspect=1)
sibsp_survived_plt.map(sns.distplot, 'SibSp')

plt.show()
#Most of survived people didn't have siblings / spouses aboard the Titanic, 
#other survived majority is people who had one sibling / spouse aboard the Titanic
```

```python
#draw a bar plot for Parch vs. survival
sns.barplot(x="Parch", y="Survived", data=train)
plt.show()
#Small families have more chance to survive, more than single.
```

```python
#draw a bar plot of survival by Pclass
sns.barplot(x="Pclass", y="Survived", data=train)

#print percentage of people by Pclass that survived
print("Percentage of Pclass = 1 who survived:", train["Survived"][train["Pclass"] == 1].value_counts(normalize = True)[1]*100)

print("Percentage of Pclass = 2 who survived:", train["Survived"][train["Pclass"] == 2].value_counts(normalize = True)[1]*100)

print("Percentage of Pclass = 3 who survived:", train["Survived"][train["Pclass"] == 3].value_counts(normalize = True)[1]*100)
```

### Cleaning the Data

```python
#we need to clear the columns which are not helping at all
data = pd.read_csv("../Dataset/titanic.csv")
data.info()
#here rows are not equal as some values are not known
# We can drop those rows, or columns
# We can substitute some values, age --> mean or median age
# Sklearn Imputer
```

```python
# Create table for missing data analysis
def find_missing_data(data):
    Total = data.isnull().sum().sort_values(ascending = False)
    Percentage = (data.isnull().sum()/data.isnull().count()).sort_values(ascending = False)
    
    return pd.concat([Total,Percentage] , axis = 1 , keys = ['Total' , 'Percent'])
```

```python
find_missing_data(data)
```

```python
# checking only train set - visualize
sns.heatmap(data.isnull(), cbar = False , 
            yticklabels = False , cmap = 'viridis')
```

```python
column_to_drop=["PassengerId","Name","Ticket","Cabin","Embarked"]
data_clean=data.drop(column_to_drop,axis=1)
data_clean.head()
```

```python
#here sex is not numerical for this we need to convert in it.
le=LabelEncoder()
data_clean["Sex"]=le.fit_transform(data_clean["Sex"])
print(data_clean.head())
print(data_clean["Fare"].loc[1])
```

```python
data_clean.info()
```

```python
#now replace age by its mean
avg_age=data_clean["Age"].mean()
data_clean=data_clean.fillna(avg_age)
data_clean.info()
```

```python
data_clean.to_csv("../Dataset/Processed data.csv",index=False)
```
